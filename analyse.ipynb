```python
import pandas as pd

# Suppose que temp_data est déjà défini et contient les données appropriées

# Liste des suffixes à traiter
suffixes = ['.1', '.2', '.3', '.4', '.5']  # Ajoute d'autres suffixes si nécessaire

# Liste pour stocker les nouveaux DataFrames
new_rows = []

# Étape 1 : Itérer sur les suffixes
for suffix in suffixes:
    # Sélectionner les colonnes avec le suffixe
    cols_with_suffix = [col for col in temp_data.columns if col.endswith(suffix)]
    
    if cols_with_suffix:
        # Créer un DataFrame avec les colonnes correspondantes
        df_suffix = temp_data[cols_with_suffix].copy()
        
        # Renommer les colonnes pour supprimer le suffixe
        df_suffix.columns = [col.replace(suffix, '') for col in df_suffix.columns]
        
        # Ajouter les nouvelles lignes à la liste
        new_rows.append(df_suffix)

result_df = pd.concat([temp_data.drop(columns=[col for col in temp_data.columns if any(s in col for s in suffixes)])] + new_rows, ignore_index=True)

# Afficher le DataFrame final
result_df.head()
```

```python
# delete les colonnes contenant Unnamed
result_df = result_df.drop(columns=[col for col in result_df.columns if 'Unnamed' in col])
```

```python
import pandas as pd

# Étape 1 : Met tout le monde en minuscule
temp_data.columns = temp_data.columns.str.lower()
cyano_data.columns = cyano_data.columns.str.lower()

temp_data = temp_data.rename(columns={'temperature': 'temperature_temp', 'date': 'date_temp'})
cyano_data = cyano_data.rename(columns={'density_cells/ml': 'density_cells_ml', 'date': 'date_cyano'})
# Étape 1 : Création d'une clé de fusion pour temp_data en utilisant 'date' et les autres colonnes
temp_data['key'] = temp_data['date_temp'].astype(str) + '_' + temp_data['reservoir'] 

# Étape 2 : Création d'une clé de fusion pour cyano_data
cyano_data['key'] = cyano_data['date'].astype(str) + '_' + cyano_data['reservoir']

# Étape 3 : Fusionner les deux DataFrames sur la clé
merged_data = pd.merge(
    cyano_data, 
    temp_data[['key', 'temperature_temp']], 
    on='key', 
    how='right'  # On garde toutes les entrées de temp_data
)

# Supprimer la colonne de clé si tu n'en as plus besoin
merged_data.drop(columns=['key'], inplace=True)

# Vérifier les données fusionnées
print(merged_data.head())

# Étape 4 : Calculer la corrélation
correlation = merged_data['temperature_temp'].corr(merged_data['density_cells/ml'])

# Afficher la corrélation
print(f"Corrélation entre la température et la densité de cyanobactéries : {correlation}")
```

```python
import pandas as pd

def fill_na_with_yearly_mean(data, year_column='Year', target_columns=[]):
    # Filtrer pour les années autres que 2011 à 2015
    filtered_data = data[~data[year_column].between(2011, 2015)]
    
    # Calculer la moyenne annuelle pour chaque colonne cible
    yearly_means = (
        filtered_data.groupby(year_column)[target_columns]
        .transform(lambda x: x.fillna(x.mean()))
    )
    
    # Remplir les valeurs manquantes dans les années hors 2011-2015 avec la moyenne annuelle
    data.update(yearly_means)
    
    return data

# Exemple d'utilisation
target_columns = ['BHR', 'BRR', 'BVR'
merged_data = fill_na_with_yearly_mean(merged_data, target_columns=target_columns)
```
